```python
y_train=np.reshape(label,(1,num_row))
```

```python
class training:
    def __init__(self):
        pass
    def train(self,x,y,no_of_iters,learn_rate,a,b): #a-num_row,b-num_col
        theta=np.zeros((1,b),dtype=float)
        self.predi=np.dot(theta,np.transpose(x))
        for m in range(0,no_of_iters):
            D=np.dot((self.predi-y),x)# 1*785
            theta=theta-((learn_rate/a)*D)
            self.predi=np.dot(theta,np.transpose(x))
        cost=(1/2*a)*np.sum(np.square(self.predi-y))
        return(theta,self.predi,cost)
    
    
    
```

```python
training().train(x_train,y_train,100,0.0000000000000001,num_row,num_col)
```
